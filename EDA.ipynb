```python
# Cell 9: Target distribution visualization
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='heart_disease')
plt.title('Distribution of Heart Disease')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Cell 10: Age distribution by heart disease
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.histplot(data=df, x='age', hue='heart_disease', kde=True)
plt.title('Age Distribution by Heart Disease')

plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='heart_disease', y='age')
plt.title('Age Boxplot by Heart Disease')
plt.tight_layout()
plt.show()
```

```python
# Cell 11: Correlation matrix
plt.figure(figsize=(12, 10))
correlation = df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.show()
```

```python
# Cell 12: Important features distribution
features = ['age', 'cholesterol', 'max_heart_rate', 'st_depression']
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.ravel()

for i, feature in enumerate(features):
    sns.boxplot(data=df, x='heart_disease', y=feature, ax=axes[i])
    axes[i].set_title(f'{feature} vs Heart Disease')

plt.tight_layout()
plt.show() 
```
